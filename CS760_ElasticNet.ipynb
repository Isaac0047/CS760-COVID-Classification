```python
import numpy as np
import scipy.io
import sklearn
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
```

```python
mat = scipy.io.loadmat('Feature_all_sig.mat')
# mat = scipy.io.loadmat('Feature_stat_tex_fftnorm')
mat['Feature_all_sig'].shape
# mat['Feature'].shape
print(mat)
```

```python
CVD     = 1252
non_CVD = 1228
Y = np.zeros((CVD+non_CVD,1))
Y[:CVD] = 1
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(mat['Feature_all_sig'], Y, test_size=0.2)
```

```python
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```python
Y_train = np.reshape(Y_train, (1984,1))
Y_test  = np.reshape(Y_test,  (496,1))
print(Y_train.shape, Y_test.shape)
```

```python
EN = ElasticNet(alpha=1.0, l1_ratio=0.5, max_iter=1000)
EN.fit(X_train, Y_train)
```

```python
EN.coef_
# EN.intercept_
```

```python
EN.predict(X_test)
```

```python
EN.predict(X_test).shape
```

```python
Y_pred = np.round(EN.predict(X_test))
Y_pred = np.reshape(Y_pred, (496,1))
```

```python
print(Y_pred.shape, Y_test.shape, (Y_pred-Y_test).shape)
```
